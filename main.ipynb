```python
import pandas as pd
from src.DB.db_config import get_db_connection

def get_users_dataframe():
    """‡∫î‡∫∂‡∫á‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô users ‡∫à‡∫≤‡∫Å DB ‡ªÅ‡∫•‡∫∞‡∫™‡∫ª‡ªà‡∫á‡∫Ñ‡∫∑‡∫ô‡ªÄ‡∫õ‡∫±‡∫ô Pandas DataFrame."""
    
    sql_query = "SELECT * FROM test_visualization.products"
    
    # ‡ªÉ‡∫ä‡ªâ 'with' statement ‡ªÄ‡∫û‡∫∑‡ªà‡∫≠‡∫à‡∫±‡∫î‡∫Å‡∫≤‡∫ô‡∫Å‡∫≤‡∫ô‡ªÄ‡∫ä‡∫∑‡ªà‡∫≠‡∫°‡∫ï‡ªç‡ªà (connection)
    try:
        with get_db_connection() as connection:
            # pd.read_sql_query() ‡∫à‡∫±‡∫î‡∫Å‡∫≤‡∫ô‡∫Å‡∫≤‡∫ô‡∫î‡∫∂‡∫á‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡ªÅ‡∫•‡∫∞‡∫õ‡ªà‡∫Ω‡∫ô‡ªÄ‡∫õ‡∫±‡∫ô DataFrame ‡ªÉ‡∫´‡ªâ‡∫ó‡ªà‡∫≤‡∫ô
            # ‡∫°‡∫±‡∫ô‡∫ç‡∫±‡∫á‡∫à‡∫±‡∫î‡∫Å‡∫≤‡∫ô‡ªÄ‡∫•‡∫∑‡ªà‡∫≠‡∫á Columns ‡ªÅ‡∫•‡∫∞ Rows ‡ªÉ‡∫´‡ªâ‡∫≠‡∫µ‡∫Å‡∫î‡ªâ‡∫ß‡∫ç
            df = pd.read_sql_query(sql_query, connection)
            return df

    except Exception as e:
        print(f"‡ªÄ‡∫Å‡∫µ‡∫î‡∫Ç‡ªç‡ªâ‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î‡ªÉ‡∫ô‡∫Å‡∫≤‡∫ô‡∫î‡∫∂‡∫á‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡∫î‡ªâ‡∫ß‡∫ç Pandas: {e}")
        return pd.DataFrame() # ‡∫™‡∫ª‡ªà‡∫á‡∫Ñ‡∫∑‡∫ô DataFrame ‡ªÄ‡∫õ‡∫ª‡ªà‡∫≤‡∫ñ‡ªâ‡∫≤‡∫°‡∫µ‡∫Ç‡ªç‡ªâ‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î

# ‡∫Å‡∫≤‡∫ô‡ªÉ‡∫ä‡ªâ‡∫á‡∫≤‡∫ô
users_df = get_users_dataframe()

# ‡∫ó‡ªà‡∫≤‡∫ô‡∫™‡∫≤‡∫°‡∫≤‡∫î‡ªÉ‡∫ä‡ªâ users_df ‡ªÄ‡∫û‡∫∑‡ªà‡∫≠‡∫ß‡∫¥‡ªÄ‡∫Ñ‡∫≤‡∫∞, ‡∫™‡∫∞‡ªÅ‡∫î‡∫á‡∫ú‡∫ª‡∫ô, ‡∫´‡∫º‡∫∑ ‡∫õ‡ªà‡∫Ω‡∫ô‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡ªÑ‡∫î‡ªâ
print(users_df.head(15))
```
